```python
import matplotlib.pyplot as plt
import numpy as np
import sys

sys.path.insert(0, '../py')
from pyxfoil import PyXFoil
```

```python
# Create PyXFoil instance
f = PyXFoil()

# Load airfoil (NACA 2412)
f.naca(2412)

# Print airfoil name
print(f.name[()].decode('utf-8'))
```

```python
# Set operating conditions
f.oper.re = 1e6
f.oper.mach = 0.3
f.oper.visc()
```

```python
# Calculate airfoil performance
cl = list()
cd = list()
cm = list()
alpha = list()
for a in np.arange(0.0, 20.0, 0.5):
    ok = f.oper.alpha(a)
    if ok:
        alpha.append(a)
        cl.append(float(f.cl))
        cd.append(float(f.cd))
        cm.append(float(f.cm))
```

```python
# Plot result
fig, ax1 = plt.subplots(figsize=(14, 6))
ax2 = ax1.twinx()
ax3 = ax1.twinx()
ax3.spines.right.set_position(('axes', 1.1))

ax1.plot(alpha, cl, color='C0')
ax2.plot(alpha, cd, color='C1')
ax3.plot(alpha, cm, color='C2')

ax1.set_xlabel('Angle of attack in ')
ax1.set_ylabel('Lift coefficient', color='C0')
ax2.set_ylabel('Drag coefficient', color='C1')
ax3.set_ylabel('Moment coefficient', color='C2')
```
